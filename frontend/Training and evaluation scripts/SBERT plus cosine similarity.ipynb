```python

import pandas as pd
from sentence_transformers import SentenceTransformer
import torch

# Load SBERT model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Load the dataset correctly
df = pd.read_csv('/content/drive/MyDrive/Diabetes_cleaned.csv')

# Make sure you select the correct text column (Change 'text_column_name' to actual column)
if 'content' not in df.columns:
    print("Error: Column 'text_column_name' not found! Check CSV headers.")
else:
    dataset = df['content'].astype(str).tolist()  # Convert to list of strings

    # Convert dataset sentences to embeddings
    dataset_embeddings = model.encode(dataset, convert_to_tensor=True)

    # Save for later use
    torch.save(dataset_embeddings, "/content/drive/MyDrive/dataset_embeddings.pt")
    torch.save(dataset, "/content/drive/MyDrive/dataset_texts.pt")

    print("Successfully stored dataset embeddings!")

```

```python
from sentence_transformers import util

# Load saved embeddings and dataset
dataset_embeddings = torch.load("/content/drive/MyDrive/dataset_embeddings.pt")
dataset = torch.load("/content/drive/MyDrive/dataset_texts.pt")

def find_most_similar_sentence(user_input, threshold=0.7):
    # Convert user input to an SBERT embedding
    query_embedding = model.encode(user_input, convert_to_tensor=True)

    # Compute cosine similarity between query and dataset
    similarities = util.pytorch_cos_sim(query_embedding, dataset_embeddings)

    # Find the most similar sentence
    max_similarity, best_match_idx = torch.max(similarities, dim=1)
    max_similarity = max_similarity.item()

    if max_similarity >= threshold:
        return f"Match found! Most similar sentence: {dataset[best_match_idx.item()]} (Similarity: {max_similarity:.2f})"
    else:
        return "No relevant match found. The query is unrelated."
```

```python

# Example Usages
print(find_most_similar_sentence("How does metformin affect blood sugar?"))
print(find_most_similar_sentence("What is the capital of France?"))
```
